```python
import pandas as pd

air_quality_file1 = "./data/PRSA_Data_Aotizhongxin_20130301-20170228.csv"
air_quality_file2 = "./data/PRSA_Data_Changping_20130301-20170228.csv"
aotizhongxin_df = pd.read_csv(air_quality_file1, header=0)
changping_df = pd.read_csv(air_quality_file2, header=0)

aotizhongxin_df.set_index('No')
```

```python
aotizhongxin_df.drop_duplicates()
print(aotizhongxin_df.shape)
```

```python
aotizhongxin_df = aotizhongxin_df.dropna()
print(aotizhongxin_df.shape)
```

```python
def c_to_f(value):
  f = 9/5 * value + 32
  return f  

aotizhongxin_df['temp_f'] = aotizhongxin_df['TEMP'].map(c_to_f)
```

```python
aotizhongxin_df['measurement_date'] = pd.to_datetime(aotizhongxin_df[['year', 'month', 'day', 'hour']])

drop_cols = ['year', 'month', 'day', 'hour']
aotizhongxin_df.drop(drop_cols, inplace=True, axis=1)
```

```python
def decode_wd(value:str):
    WD_MAP = {
        'N': 'North', 
        'NNE': 'North Northeast', 
        'NE': 'Northeast',
        'ENE': 'West', 
        'E': 'East',
        'ESE': 'East Southeast',
        'SE': 'Southeast',
        'SSE': 'South Southeast',
        'S': 'South',
        'SSW': 'South Southwest',
        'SW': 'Southwest',
        'WSW': 'West Southwest',
        'W': 'West',
        'WNW': 'West Northwest',
        'NW': 'Northwest',
        'NNW': 'North Northwest',

    }
    if value in WD_MAP:
        return WD_MAP[value]
    else:
      return value

aotizhongxin_df['wd'] = aotizhongxin_df.wd.map(decode_wd)
```
